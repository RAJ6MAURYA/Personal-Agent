```python
import openai
import pprint
```

```python
client = openai.OpenAI(base_url="http://localhost:11434/v1",api_key="asdf")
```

```python
stream = client.chat.completions.create(model="llama3.2",messages=[{"role":"user","content":"Hi my name is raj"}],stream=True)

for chunk in stream:
    delta = chunk.choices[0].delta.content
    if delta:
        print(f'{delta}',flush=True,end="")
```
